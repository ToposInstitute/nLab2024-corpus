```python
import os 
import re
from bidict import bidict
#from difflib import SequenceMatcher as SM
from tqdm import tqdm

wd = {}
nl = {}

with open('nlab_page_names_raw.txt', 'r', encoding="utf-8") as input:
    cnt = 0
    for cnt in range(19400):
        item = input.readline().lower().strip()
        if item == "":
            break
        if not " > history" in item:
            nl[item] = True

with open('wiki.txt', 'r', encoding="utf-8") as input:
    cnt = 0
    for cnt in tqdm(range(100000000), desc="Loading..."):
        rec = input.readline()
        if rec == "":
            break
        key = rec.split(",")[1].strip()
        value = rec.split(",")[0].strip().lower()        
        wd[key] = value
```

```python
cnt = 0
for(key, value) in wd.items():
    cnt += 1
    if key in nl:
        print(key)
```
